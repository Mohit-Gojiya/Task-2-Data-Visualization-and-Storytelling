```python
import pandas as pd
import numpy as np

# Load data
df = pd.read_csv("dirty_cafe_sales.csv")

# See top 20 rows
df.head(20)
```

```python
#Clean Column Names (Transaction ID ---> transaction_id)

df.columns = df.columns.str.lower().str.strip().str.replace(" ","_")
df.head()
```

```python
#Replace Invalid Entries
df.replace(["ERROR", "UNKNOWN", "None", ""], np.nan, inplace=True)
df.head(20)
```

```python
#check unwanted values

for col in df.columns:
    print(f"\nColumn: {col}")
    print(df[col].value_counts(dropna=False))
```

```python
#Handle Missing Values

#Option 1: Drop rows with too many NaNs (* keep rows with at least 5 non-null values)
df.dropna(thresh=5, inplace=True)  
df.info()
```

```python
#Check Missing Values
df.isnull().sum().sort_values(ascending=False)
```

```python
# Fill missing values in 'item' column with 'Unknown' to handle invalid or missing entries
df['item'] = df['item'].fillna('Unknown')

# Replace missing values in 'quantity' column with the median value for more robust imputation
df['quantity'] = df['quantity'].fillna(df['quantity'].median())

# Replace missing values in 'price_per_unit' with its median — prevents outlier skew compared to mean
df['price_per_unit'] = df['price_per_unit'].fillna(df['price_per_unit'].median())

# Fill missing payment methods with the most common value (mode) to maintain category distribution
df['payment_method'] = df['payment_method'].fillna(df['payment_method'].mode()[0])

# Fill missing locations with the most frequent one to preserve distribution
df['location'] = df['location'].fillna(df['location'].mode()[0])

# Forward-fill missing transaction dates — assumes chronological order; best for time-series
df['transaction_date'] = df['transaction_date'].ffill()

# Check for remaining missing values after handling
df.isnull().sum()
```

```python
#Fix Incorrect Data Types

# View current data types of all columns
df.dtypes
```

```python
# Convert quantity to integer (round first if needed, then cast)
df['quantity'] = df['quantity'].round().astype('int')

# Convert transaction_id, item, payment_method, location to 'category' (saves memory and helps with modeling)
df['transaction_id'] = df['transaction_id'].astype('category')
df['item'] = df['item'].astype('category')
df['payment_method'] = df['payment_method'].astype('category')
df['location'] = df['location'].astype('category')

# price_per_unit and total_spent are already float64 — no change needed
# transaction_date is already datetime — no change needed

# Final check
df.info()
```

```python
df.to_csv("cleaned_coffee_sales.csv", index=False)
```
